```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import os
import scipy.optimize as opt
```

## 2A

```python
def linear(x, a):
    return a * x

def select_y_range(x, y, minimum, maximum):
    region = np.logical_and((y > minimum), (y < maximum))
    yy = y[region]
    xx = x[region]
    return (xx, yy)
```

```python
data_2a = np.genfromtxt('data/2A.txt', skip_header=True)
I1, V1, V2 = data_2a.T
ΔV = V1 - V2
I_sel, V_sel = select_y_range(I1, ΔV, -1, 1)
result, cov = opt.curve_fit(linear, I_sel, V_sel)
x_range = np.linspace(-1, 1)
r = result[0]

plt.plot(V_sel, I_sel, label='measured')
plt.plot(x_range, linear(x_range, 1/r), linestyle=':', color='grey', label=f'$ΔV={r:.2f} \\cdot I$')
plt.legend()

plt.ylabel('I (A)')
plt.xlabel('ΔV (V)')
#plt.savefig('images/2a.jpg', dpi=720, bbox_inches='tight')
```

## 8A

```python
Vgs = [0, 1, 2, 3, 4, 5]
Vbs = [0, -1, -2]
Vds = [5, 10, 15]

def splitter(volts, curr):
    splitvolts = []
    splitcurrs = []
    begin = 0
    for i, v in enumerate(volts[:-1]):
        if volts[i+1] < v:
            splitvolts.append(volts[begin:i+1])
            splitcurrs.append(curr[begin:i+1])
            begin = i+1
    splitvolts.append(volts[begin:])
    splitcurrs.append(curr[begin:])
    return splitvolts, splitcurrs


def plot_vg_splitter(filename):
    data = np.genfromtxt(filename, skip_header=True)
    Vs, Is = splitter(*data.T)
    for (v, i, vg) in zip(Vs, Is, Vgs):
        plt.plot(v, i, label=f'$V_g=${vg}V')
    plt.legend()
    plt.xlabel('$V_d$ (V)')
    plt.ylabel('$I_d$ (A)')
    base = os.path.basename(filename)[:-4]
    #plt.savefig('images/' + base + '.jpg', dpi=720, bbox_inches='tight')
    return

def plot_vb_splitter(filename):
    data = np.genfromtxt(filename, skip_header=True)
    Vs, Is = splitter(*data.T)
    for (v, i, vb) in zip(Vs, Is, Vbs):
        plt.plot(v, i, label=f'$V_b=${vb}V')
    plt.legend()
    plt.xlabel('$V_g$ (V)')
    plt.ylabel('$I_d$ (A)')
    base = os.path.basename(filename)[:-4]
    #plt.savefig('images/' + base + '.jpg', dpi=720, bbox_inches='tight')
    plt.show()
    
    # clear figure for log plot
    plt.clf()
    for (v, i, vb) in zip(Vs, Is, Vbs):
        plt.plot(v, np.log10(i), label=f'$V_b=${vb}V')
    plt.legend()
    plt.xlabel('$V_g$ (V)')
    plt.ylabel('$\\log I_d$ (A)')
    
    #plt.savefig('images/' + base + '_log.jpg', dpi=720, bbox_inches='tight')
    plt.show()
    return

def plot_inverter_splitter(filename):
    data = np.genfromtxt(filename, skip_header=True)
    Vs, Is = splitter(*data.T)
    for (v, i, vd) in zip(Vs, Is, Vds):
        plt.plot(v, i, label=f'$V_g=${vd}V')
    plt.legend()
    plt.xlabel('$V_{in}$ (V)')
    plt.ylabel('$V_{out}$ (V)')
    base = os.path.basename(filename)[:-4]
    #plt.savefig('images/' + base + '.jpg', dpi=720, bbox_inches='tight')
    return
```
